```python
def get_product_len(rxn_smart):
    try:
        product_len = len(AllChem.ReactionFromSmarts(rxn_smart).GetProducts())
        if product_len > 3:
            print(f"{product_len} product: ", rxn_smart)
        return product_len
    except:
        print(rxn_smart)
        return 0
```
